```python
def parse_data(file):
    """
    Parse saved csv data file each column as individual arrays
    
    Parameters
    ----------
    file: string
        string to the datafile pathname
    """

    data = np.loadtxt(file, delimiter=",")
    data = np.array([data[:, i] for i in range(data.shape[1])])
    return data
```

## Linear and Expotnetial Decay Regression

```python
sp30_max = np.argmax(data6[4])
sp45_max = np.argmax(data7[4])

print(sp30_max)
print(data6[4][sp30_max])

print(sp45_max)
print(data7[4][sp45_max])
```

```python
from scipy.optimize import curve_fit

def exp_decay(t, y0, A, k):
    return y0 + A * np.exp(k * t)

def lin_decay(t, y0, m):
    return y0 + m * t

def cutoff(t, y, start_idx, slope_means, slope_stds, snrs, box_size=50, epsilon=10**-12, step=1, snr_cutoff=1.0, slope_cutoff=0.05, method='snr'):
    """
    Determine cutoff of decaying plot with moving average with
    slope as criteria for cut-off
    
    Parameters
    ----------
    t: np.array
        time array
    y: np.array
        CO2 concentration array
    start_idx: int
        index of the max temeprature value, starting point 
        of regression
    box_size: int
        size of moving box
    snr: float
        signal to noise ratio cutoff for regression slope
    """

    n = len(y) # len of data from 
    consecative = 0

    for i in range(start_idx, n - box_size, step): # start from start_idx + box center, end likewise
        box = slice(i, i + box_size)
        t_box = t[box]
        y_box = y[box]

        if method == 'snr':
            # m, b = np.polyfit(t_box, y_box, 1) # linear regression
            m = np.cov(t_box, y_box, ddof=1)[0, 1] / np.var(t_box, ddof=1) # linear regression in terms of covariance and variance
            slopes = np.diff(y_box) / np.diff(t_box) # array of numerical forward differences
            slope_mean = abs(m) # average slope from regression 
            slope_std = np.std(slopes, ddof=1)
            
            if slope_std < epsilon:
                snr = np.inf
            else:
                snr = slope_mean / slope_std

            
            slope_means.append(m)
            slope_stds.append(slope_std)
            snrs.append(snr)

            if snr < snr_cutoff:
                consecative += 1
            else: 
                consecative = 0
            
            if consecative >= 10:
                return i + box_size
            
        if method == 'slope':
            m = np.cov(t_box, y_box, ddof=1)[0, 1] / np.var(t_box, ddof=1)
            slope_mean = abs(m)

            if slope_mean < slope_cutoff:
                consecative += 1
            else: 
                consecative = 0

            if consecative >= 10:
                return i + box_size
        
    return None
```

## Individual Plots

```python
t_window = data6[0][387:702]
y_window = data6[4][387:702]
t_rel = t_window - t_window[0]

popt, pcov = curve_fit(exp_decay, t_rel, y_window, p0=[y_window[-1], y_window[0]-y_window[-1], -0.01])
popt_lin, _ = curve_fit(lin_decay, t_rel, y_window)

y_fit = exp_decay(t_rel, *popt)
residuals = y_window - y_fit
ss_res = np.sum(residuals**2)
ss_tot = np.sum((y_window - np.mean(y_window))**2)
r2 = 1 - (ss_res / ss_tot)

y_fit_lin = lin_decay(t_rel, *popt_lin)
r2_lin = 1 - np.sum((y_window - y_fit_lin)**2) / np.sum((y_window - np.mean(y_window))**2)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(t_window, y_window, label=label_array[3])
ax.plot(t_window, exp_decay(t_rel, *popt), '--k', label=f'exp fit, k={popt[2]:.4f}, r^2={r2:.4f}')
ax.plot(t_window, lin_decay(t_rel, *popt_lin), ':k', label=f'linear fit, m={popt_lin[1]:.4f}, r^2={r2_lin:.4f}')
ax.set_xlabel('Time (s)')
ax.set_ylabel(y_axis_labels[1])
ax.tick_params(axis='y')
ax.legend()
ax.grid(True)
```
